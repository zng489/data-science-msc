```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
In11130500 = pd.read_csv('In11-130-500.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
In8110500 = pd.read_csv('In8-110-500.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
In7120650 = pd.read_csv('In7-120-650.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = plt.subplot(231)
In11130500.plot(x='weight',y='conc', ax=ax1,label='In-130°C-1:500')
plt.xlabel('')
plt.ylabel('')
#red_patch = mpatches.Patch(color='red', label='The red data')
#plt.legend(loc='upper left', frameon=False)
plt.legend('')
ax1.set_title('In-130°C-1:500')

ax2 = plt.subplot(232)
In7120650.plot(x='weight',y='conc', ax=ax2, label='In-120°C-1:650')
plt.xlabel('')
plt.ylabel('')
plt.legend('')
ax2.set_title('In-120°C-1:650')

ax3 = plt.subplot(233)
In8110500.plot(x='weight',y='conc', ax=ax3, label='In-110°C-1:500')
plt.xlabel('')
plt.ylabel('')
plt.legend('')
ax3.set_title('In-110°C-1:500')


ax4 = plt.subplot(212)
ax4.set_title('')
ax4.label_outer()
In11130500.plot(x='weight',y='conc', ax=ax4, label='In-130°C-1:500')
In7120650.plot(x='weight',y='conc', ax=ax4, label='In-120°C-1:650')
In8110500.plot(x='weight',y='conc', ax=ax4, label='In-110°C-1:500')
plt.xlabel('Log(Mw)')
plt.ylabel('d(wt)/d[log(Mw)]')
plt.legend(title='$BMI.In_{2}Cl_{7}$')

plt.savefig("GPCbmiin2cl7.jpg", format='jpg', dpi=1200)

plt.tight_layout()
plt.show()
```
