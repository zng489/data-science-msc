```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
Cu1D130500 = pd.read_csv('Cu1D130500.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
Cu1E120650 = pd.read_csv('Cu1E120650.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
Cu1A110500 = pd.read_csv('Cu1A110500.txt', delimiter='\t', names=['1','2','3','weight','5','6','7','8','9','conc'])
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = plt.subplot(231)
Cu1D130500.plot(x='weight',y='conc', ax=ax1,label='Cu-130°C-1:500')
plt.xlabel('')
plt.ylabel('')
#red_patch = mpatches.Patch(color='red', label='The red data')
#plt.legend(loc='upper left', frameon=False)
plt.legend('')
ax1.set_title('Cu-130°C-1:500')

ax2 = plt.subplot(232)
Cu1E120650.plot(x='weight',y='conc', ax=ax2, label='Cu-120°C-1:650')
plt.xlabel('')
plt.ylabel('')
plt.legend('')
ax2.set_title('Cu-120°C-1:650')

ax3 = plt.subplot(233)
Cu1A110500.plot(x='weight',y='conc', ax=ax3, label='Cu-110°C-1:500')
plt.xlabel('')
plt.ylabel('')
plt.legend('')
ax3.set_title('Cu-110°C-1:500')


ax4 = plt.subplot(212)
ax4.set_title('')
ax4.label_outer()
Cu1D130500.plot(x='weight',y='conc', ax=ax4, label='Cu-130°C-1:500')
Cu1E120650.plot(x='weight',y='conc', ax=ax4, label='Cu-120°C-1:650')
Cu1A110500.plot(x='weight',y='conc', ax=ax4, label='Cu-110°C-1:500')
plt.xlabel('Log(Mw)')
plt.ylabel('d(wt)/d[log(Mw)]')
plt.legend(title='$BMI.Cu_{2}Cl_{3}$')

plt.savefig("GPCbmicu2cl3.jpg", format='jpg', dpi=1200)

plt.tight_layout()
plt.show()

```
