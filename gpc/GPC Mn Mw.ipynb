```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
gpc = pd.read_excel('gpc.xlsx', names=['catalisador','mn','mw'])
```

```python
gpc
```

```python
gpc['mw']
```

```python
plt.figure(figsize=(20,16))

plt.barh(gpc['catalisador'], gpc['mn'], label='')
plt.ylabel('Catalisadores', size= 30)
plt.xlabel('Peso molecular númerica (Mn)', size= 30)
#red_patch = mpatches.Patch(color='red', label='The red data')
#plt.legend(loc='upper left', frameon=False)
plt.legend('')
plt.xticks(np.arange(0,2000,step=200),size= 20) #0.5)
plt.yticks(size= 20) #0.5)



#plt.yticks(np.arange(0,3200,step=200), size= 20) #0.5))
#plt.set_title('Conversão dos catalisadores')
#plt.xticks(rotation=90)

plt.savefig("mn.jpg", format='jpg', dpi=1200)


plt.tight_layout()
plt.show()
```
